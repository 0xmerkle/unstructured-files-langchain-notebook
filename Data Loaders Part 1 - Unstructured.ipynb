### Load Documents

```python
doc_loader = UnstructuredFileLoader("./data/rex-woodbury-substacks/retail_revolution.txt")
doc = doc_loader.load()
```

```python
def summary(docs):
    print (f'You have {len(docs)} document(s)')
    
    num_words = sum([len(doc.page_content.split(' ')) for doc in docs])
    
    print (f'You have roughly {num_words} words in your docs')
    print ()
    print (f'Preview: \n{docs[0].page_content.split(". ")[0]}')
```

```python
summary(doc)
```

### Load Your LLM

```python
from langchain import OpenAI
```

```python
OPENAI_API_KEY = '...'
```

```python
llm = OpenAI(openai_api_key=OPENAI_API_KEY)
```

### Summarize: Stuff

```python
print(doc)
```

```python
chain = load_summarize_chain(llm, chain_type="map_reduce", verbose=True)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size = 400,
    chunk_overlap = 0
)
```

```python
chunks = text_splitter.split_documents(doc)
```

```python
summary(chunks)
```

```python
chain.run(chunks)
```

### Q&A

```python
chain = load_qa_chain(llm, chain_type="map_reduce", verbose=True, return_intermediate_steps=True)
```

```python
summary(chunks)
```

```python
query = "what are One of the challenges of online shopping?"

answer = chain({"input_documents": chunks, "question": query}, return_only_outputs=True)
```

```python
print(answer['output_text'])
```
